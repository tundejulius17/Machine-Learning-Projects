### Exploring the performance of machine learning models with feature selection.

```python
from scipy.io import loadmat
import numpy as np
```

```python
mat_data = loadmat('arcene.mat')
```

```python
mat_data.keys()
```

```python
X_train = mat_data['X_train']
X_test = mat_data['X_test']
```

```python
X_train.shape, X_test.shape
```

```python
X_train
```

```python
y_train = mat_data['y_train'].ravel()
y_test = mat_data['y_test'].ravel()
```

```python
y_train.shape, y_test.shape
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, random_state=0)
classifier.fit(X_train, y_train)
```

```python
feature_importances = classifier.feature_importances_
```

```python
feature_importances
```

```python
len(feature_importances) # number of features
```

```python
np.count_nonzero(feature_importances)
```

```python
# plot of the feature importances
import matplotlib.pyplot as plt
plt.plot(feature_importances)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importances Plot')
```

```python
y_pred = classifier.predict(X_test) # prediction
```

```python
# accuracy
from sklearn.metrics import accuracy_score
print('Prediction accuracy: {:.2f} %'.format(accuracy_score(y_test, y_pred) * 100))
```

```python
from sklearn.feature_selection import RFECV
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
linear_discriminant_clf = LinearDiscriminantAnalysis()
rfe = RFECV(estimator=linear_discriminant_clf, step= 50, verbose=1) #RFECV constructor
```

```python
rfe.fit(X_train, y_train) # fit rfe
```

```python
selected_features = rfe.support_ # get selected features
```

```python
print('Number of selected features: ', np.count_nonzero(selected_features))
```

```python
# plot errors for the features
plt.plot(range(0,10001,50), rfe.grid_scores_)
plt.title('Plot of errors for the different features')
plt.xlabel('Features')
plt.ylabel('CV score of i-th subset of features')
```

```python
y_pred_rfe = rfe.predict(X_test) # predict using rfe
```

```python
print('Prediction accuracy for RFE: {:.2f} %'.format(accuracy_score(y_test, y_pred_rfe) * 100))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
logistic_regression_clf = LogisticRegression(penalty='l1', solver='liblinear')
```

```python
C_values = 10.0 ** np.arange(-4, 2, 1)
```

```python
C_values.shape
```

```python
accuracy_scores_dict = {}
for C in C_values:
  logistic_regression_clf.C = C
  C_score = cross_val_score(logistic_regression_clf, X_train, y_train, cv=5).mean()
  print("Validation accuracy is {:.2f} % for C = {}".format(C_score*100, logistic_regression_clf.C))
  accuracy_scores_dict[C] = C_score*100

best_C_value = max(accuracy_scores_dict, key=accuracy_scores_dict.get)
print('Best C value is', best_C_value)
```

```python
logistic_regression_clf.C = best_C_value
logistic_regression_clf.fit(X_train, y_train)
```

```python
selected_features_clf = np.count_nonzero(logistic_regression_clf.coef_)
print('The number of selected features is:',selected_features_clf)
```

```python
y_pred_best_C = logistic_regression_clf.predict(X_test)
```

```python
print('Prediction accuracy: {:.2f} %'.format(accuracy_score(y_test, y_pred_best_C) * 100))
```
